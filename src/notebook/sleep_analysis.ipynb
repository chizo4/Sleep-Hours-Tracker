```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3
```

```python
# Connect with the DB and create a dataframe for this purpose.
db = sqlite3.connect('../slept_hours.db')
df = pd.read_sql_query("SELECT * FROM sleep_table", db)
```

```python
# Get the 3 oldest DB records.
df.head(3)
```

```python
# Get the 3 latest DB records.
df.tail(3)
```

```python
# Plotting scatter diagram for the day-by-day data for the current year (2022).
regex = '\d{2}[/]\d{2}[/]2022'
df_2022 = df[df.date.str.match(regex)]

plt.figure(dpi=200)
#df_2022.plot.scatter(x='date', y='hours', color='blue')

# create data
x = df_2022['date']
y = df_2022['hours']

# set title
plt.title('sleep hours 2022 day-by-day')
plt.ylabel('slept hours (h)')
plt.xlabel('date')
  
# plot lines
plt.plot(x, y, label = "hours slept (per day)")

plt.ylim((5.0,10.0))
#plt.xticks([])
plt.xticks(np.arange(0, len(x)+1, 10))
plt.xticks(fontsize=8, rotation=90)

plt.legend()
plt.show()
```
